#Import libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

#Load dataset

```python
from sklearn.datasets import load_digits

dataset = load_digits()
dataset.data
```

# Data Preparing

```python
X, y = dataset.data, dataset.target
```

```python
dataset.target[:40]
```

```python
begin_data = dataset.data[:30]
begin_data
```

# Hierarchy Clustering Method

## Single linkage

```python
from scipy.cluster.hierarchy import dendrogram, linkage

links_single = linkage(begin_data, 'single')

plt.figure(figsize=(19, 8))

dendrogram(links_single)

plt.xlabel('Row')
plt.ylabel('Distance')
plt.title('Method = single')

plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(8, 8))

axes[0].imshow(X[0].reshape(8, 8), cmap='gray')
axes[0].set_title('Image 0')

axes[1].imshow(X[10].reshape(8, 8), cmap='gray')
axes[1].set_title('Image 10')

axes[2].imshow(X[20].reshape(8, 8), cmap='gray')
axes[2].set_title('Image 20')

plt.tight_layout()
plt.show()
```

## Complete linkage

```python
links_complete = linkage(begin_data, method='complete')

plt.figure(figsize=(19, 8))

dendrogram(links_complete)

plt.xlabel('Row')
plt.ylabel('Distance')
plt.title('Method = complete')
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(8, 8))

axes[0].imshow(X[7].reshape(8, 8), cmap='gray')
axes[0].set_title('Image 7')

axes[1].imshow(X[17].reshape(8, 8), cmap='gray')
axes[1].set_title('Image 17')

axes[2].imshow(X[27].reshape(8, 8), cmap='gray')
axes[2].set_title('Image 27')

plt.tight_layout()
plt.show()
```

# Sklearn Agglomerative

## Using PCA (n_components = 2)

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
X_pca.shape
```

```python
pca.explained_variance_ratio_
```

```python
pca.explained_variance_ratio_.sum()
```

```python
pca.n_components_
```

### Single linkage

```python
from sklearn.cluster import AgglomerativeClustering

groups_single = AgglomerativeClustering(n_clusters=10, metric='manhattan', linkage='single', compute_distances=True)

groups_single.fit_predict(X_pca)
```

```python
plt.figure(figsize=(17, 10))

for label in dataset.target_names:
  plt.scatter(X_pca[groups_single.labels_ == label, 0],
                  X_pca[groups_single.labels_ == label, 1],
                  label=label)

# plt.scatter(*X_pca.T, c=groups_single.labels_, cmap='viridis')
# plt.colorbar()
plt.title('linkage = single')
plt.legend(loc='upper left')
plt.show()
```

```python
unique, counts = np.unique(groups_single.labels_, return_counts=True)

print(np.asarray((unique, counts)).T)
```

```python
groups_complete = AgglomerativeClustering(n_clusters=10, metric='manhattan', linkage='complete', compute_distances=True)

groups_complete.fit_predict(X_pca)
```
